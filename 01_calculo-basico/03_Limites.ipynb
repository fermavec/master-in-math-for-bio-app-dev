# Concepto de Límite

En matemáticas, el **límite** de una función describe el comportamiento de la función a medida que su argumento se aproxima a un valor específico. Es una herramienta fundamental en el análisis matemático.

## Notación del Límite
Si una función \( f(x) \) se aproxima a un valor \( L \) a medida que \( x \) se aproxima a un valor \( a \), se escribe como:

$$
\lim_{{x \to a}} f(x) = L
$$

Esto se lee como "el límite de \( f(x) \) cuando \( x \) tiende a \( a \) es igual a \( L \)".

# Tipos de Límite

## Límite Finito en un Punto Finito
Cuando \( x \) se aproxima a un valor \( a \) y el límite de \( f(x) \) es un número real \( L \):

$$
\lim_{{x \to a}} f(x) = L
$$

## Límite Infinito en un Punto Finito
Cuando \( x \) se aproxima a un valor \( a \) y \( f(x) \) crece sin límite positivo o negativo:

$$
\lim_{{x \to a}} f(x) = \infty
$$

$$
\lim_{{x \to a}} f(x) = -\infty
$$

## Límite Finito en el Infinito
Cuando \( x \) crece sin límite positivo o negativo y \( f(x) \) se aproxima a un número real \( L \):

$$
\lim_{{x \to \infty}} f(x) = L
$$

$$
\lim_{{x \to -\infty}} f(x) = L
$$

## Límite Infinito en el Infinito
Cuando \( x \) crece sin límite positivo o negativo y \( f(x) \) también crece sin límite positivo o negativo:

$$
\lim_{{x \to \infty}} f(x) = \infty
$$

$$
\lim_{{x \to \infty}} f(x) = -\infty
$$

$$
\lim_{{x \to -\infty}} f(x) = \infty
$$

$$
\lim_{{x \to -\infty}} f(x) = -\infty
$$

## Límite por la Derecha e Izquierda
Cuando se considera el comportamiento de \( f(x) \) a medida que \( x \) se aproxima a \( a \) desde la derecha (valores mayores) o desde la izquierda (valores menores):

- **Límite por la derecha**:

$$
\lim_{{x \to a^+}} f(x) = L
$$

- **Límite por la izquierda**:

$$
\lim_{{x \to a^-}} f(x) = L
$$

# Ejemplos de Límites

1. 
$$
\lim_{{x \to 2}} (3x + 1) = 7
$$
2. 
$$
\lim_{{x \to 0}} \frac{1}{x^2} = \infty
$$
3. 
$$
\lim_{{x \to \infty}} \frac{1}{x} = 0
$$

Los límites son esenciales para definir conceptos como la derivada y la integral en el cálculo.
